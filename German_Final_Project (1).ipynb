### Train and test our ModelðŸ¤·â€â™‚ï¸

```python
x_train, x_test, y_train, y_test = train_test_split(    
 
    x, y, test_size=0.2,stratify=y,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
def train_model(model,param_grid,x_train,y_train,x_test,y_test):
    grid_search = GridSearchCV(estimator=model,
                               param_grid=param_grid,
                               scoring='accuracy',
                               cv=5,
                               n_jobs=-1)
    
    grid_search.fit(x_train, y_train)
    
    best_model = grid_search.best_estimator_
    
    y_pred = best_model.predict(x_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    
    return best_model, accuracy, grid_search.best_params_
```

```python
dt= DecisionTreeClassifier(random_state=42,
                           class_weight='balanced')
dt_param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
best_dt, dt_accuracy, dt_best_params = train_model(
    dt, dt_param_grid, x_train, y_train, x_test, y_test)
```

```python
print("Best Decision Tree Model:", best_dt)
print("Decision Tree Accuracy:", dt_accuracy)
```

```python
rf= RandomForestClassifier(random_state=42,
                           class_weight='balanced',
                           n_jobs=-1)
rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
best_rf, rf_accuracy, rf_best_params = train_model(
    rf, rf_param_grid, x_train, y_train, x_test, y_test)

print("Best Random Forest Model:", best_rf)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
et= ExtraTreesClassifier(random_state=42,
                         class_weight='balanced',
                            n_jobs=-1)

et_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5,10],
    'min_samples_leaf': [1, 2, 4]
}
best_et, et_accuracy, et_best_params = train_model(
    et, et_param_grid, x_train, y_train, x_test, y_test)
print("Best Extra Trees Model:", best_et)
print("Extra Trees Accuracy:", et_accuracy)
```

```python
numeric_cols=['Age','Job','Duration']
```

```python
def detect_outliers_std(German_credit, col, threshold=3):
    mean = German_credit[col].mean()
    std = German_credit[col].std()
    lower_bound = mean - threshold * std
    upper_bound = mean + threshold * std
    outliers = German_credit[(German_credit[col] < lower_bound) | (German_credit[col] > upper_bound)]
    return outliers
```

```python
for col in numeric_cols:
    outliers = detect_outliers_std(df_credit,col)
    print(f"\nOutliers in '{col}' using Standard Deviation:")
    print(outliers[[col]])
```

### Correlation Analysis

##### The correlation matrix above shows the relationships among the numeric variables in the dataset. There is a strong positive correlation between Credit amount and Duration (r = 0.62), indicating that higher credit amounts are generally associated with longer loan durations.

##### Job also shows a moderate positive correlation with both Credit amount (r = 0.29) and Duration (r = 0.21), suggesting that employment type may influence the amount and period of credit issued. 

##### In contrast, Age has very weak correlations with all other variables, implying it has little linear relationship with credit amount, job type, or duration.

#### From the analysis, a few interesting cases were selected for presentation:

##### High credit amount with long duration â€“ representing clients with potentially higher financial risk.

##### Low savings but high credit request â€“ showing possible inconsistency between financial capacity and credit needs.

##### Young applicants with short duration loans â€“ highlighting low-risk short-term borrowers.

##### Cases with missing saving/checking information â€“ illustrating the impact of incomplete financial records.

#### C. Technique Selection and Justification

EDA is essential in identifying data quality issues, uncovering hidden patterns, detecting outliers, and understanding variable relationships.

##### Additionally, correlation analysis and visual analytics (using scatter plots, heatmaps, and boxplots) were applied to quantify and visualize relationships among variables. These methods are appropriate since they:

##### Handle mixed data types (numerical and categorical).

##### Provide both statistical and visual insights.

##### If predictive modeling were to follow, the insights gained from EDA would guide the feature selection and model choice (e.g., logistic regression or random forest for credit risk classification).
